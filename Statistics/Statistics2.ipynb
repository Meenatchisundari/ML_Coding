```python
# Z Score
outliers=[]

def detect_outliers(data):
  threshold=2 ## 2 std deviation
  mean=np.mean(data)
  std=np.std(data)

  for i in data:
    z_score=(i-mean)/std
    if np.abs(z_score)> threshold:
      outliers.append(i)

  return outliers

```

```python
import numpy as np



outliers = []

def detect_outliers(data):
    threshold = 3  # 3 standard deviations
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)

    return outliers

print(detect_outliers(dataset))
```

```python
#IQR
'''1.SORT THE DATA
2.Calculate Q1 and Q3
3.IQR = Q3-Q1
4.Find the lower fence = Q1-1.5*IQR
5.Find the upper fence = Q3+1.5*IQR
'''
dataset=sorted(dataset)
dataset
```

```python
q1,q3=np.percentile(dataset,[25,75])
```

```python
q1,q3
```

```python
iqr = q3-q1
print(iqr)
```

```python
lower_fence = q1-(1.5*iqr)
upper_fence = q3+(1.5*iqr)
print(lower_fence,upper_fence)
```
