```python
import numpy as np

class LinearRegressionNormalEq:
  def __init__(self):
    self.w = None #[b,w1,w2,w3..]

  def fit(self,X,y):
    X = np.asarray(X, dtype=float)
    y = np.asarray(y, dtype=float)

    n_samples = X.shape[0]

    #Add bias column (1s)
    X_bias = np.c_[np.ones(n_samples), X]

    #Normal equation theta =(X_transpose.X)^-1 (X_transpose.y)
    X_transpose = X_bias.T
    X_transpose_X = X_transpose @ X_bias
    X_transpose_y = X_transpose @ y
    self.w = np.linalg.inv(X_transpose_X) @ X_transpose_y


    """  #Noraml eqn (stable):
    self.w = np.linalg.pinv(X_bias)@y
    """
    return self

  def predict(self,X):
    X = np.asarray(X, dtype=float)
    n_samples = X.shape[0]
    X_bias = np.c_[np.ones(n_samples),X]
    return X_bias @ self.w
```

```python
X = [[1], [2], [3], [4]]
y = [3, 5, 7, 9]

m = LinearRegressionNormalEq().fit(X, y)
print(m.w)             # should be close to [1, 2]
print(m.predict([[5]])) # should be 11
```
