```python
import numpy as np

#x values
X =np.array([1,2,3,4],dtype=float)


#y=2x+1
y=np.array([3,5,7,9], dtype=float)

w=0.00 #slope
b=0.00 #intercept
lr=0.1 #learning rate

#Prediction
y_pred = w*X +b
print("y_pred:",y_pred)

#MSE
error = y_pred -y
loss = np.mean(error**2)
print("loss:",loss)

#Gradients(find direction to move)
dw = 2*np.mean(error*X)
db = 2*np.mean(error)

#Update the slope
w=w-lr*dw
b=b-lr*db

print("w:",w)
print("b:",b)


#Gradient descent(loop)
for i in range(1000):
  y_pred = w*X+b
  error=y_pred-y
  dw = 2*np.mean(error*X)
  db = 2*np.mean(error)

  w=w-lr*dw
  b=b-lr*db

print("w:",w)
print("b:",b)
print("loss:",np.mean(error**2))
print("y_pred:",y_pred)
```

w → vector (one weight per feature)

b → scalar (one intercept)

X shape = (n, d)

w shape = (d,)

```python
import numpy as np

class LinearRegressionGD:
    def __init__(self, lr=0.01, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        # Convert to numpy arrays
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float).reshape(-1)

        n_samples, n_features = X.shape

        # Initialize parameters
        self.w = np.zeros(n_features)
        self.b = 0.0

        # Gradient Descent
        for _ in range(self.n_iters):
            # Prediction
            y_pred = X @ self.w + self.b

            # Error
            error = y_pred - y

            # Gradients
            dw = (2 / n_samples) * (X.T @ error)
            db = (2 / n_samples) * np.sum(error)

            # Update
            self.w -= self.lr * dw
            self.b -= self.lr * db

        return self

    def predict(self, X):
        X = np.asarray(X, dtype=float)
        return X @ self.w + self.b
```

```python
X = [
    [1, 2],
    [2, 1],
    [3, 4],
    [4, 3]
]

# y = 3*x1 + 2*x2 + 1
y = [8, 9, 18, 19]

model = LinearRegressionGD(lr=0.01, n_iters=2000)
model.fit(X, y)

print("Weights:", model.w)
print("Bias:", model.b)

# New input
X_new = [[5, 2]]
print("Prediction:", model.predict(X_new))
```
